## Explore data

We open up the data and check the data type, shape, and min and max 

```python
from pathlib import Path
from skimage.io import imread

import sys
sys.path.append('../')
import helper

#parent_path = r"../../data/bees/images"
parent_path = r"../../data/roots/images"
```

```python
# list files in parent path
parent_path = Path(parent_path)
files = list(parent_path.glob('*.tif'))
files = files+list(parent_path.glob('*.jpg'))

for file in files:
    img = imread(file)
    print(file)
    print(img.shape)
    print(img.dtype)
    print(img.min(), img.max())

    print()
```
